```python
# Write a Pandas function to drop rows that contain a missing value.
import pandas as pd
import numpy as np

def drop_missing_rows():
    df = pd.DataFrame(iris_2d)
    df_clean = df.dropna()
    return df_clean
print(drop_missing_rows())
```

```python
# Write a Pandas function to find the mean, median and standard deviation of column 3.
def mean_column_3():
    df = pd.DataFrame(iris_2d)
    column = df[3]   # select column 3 (0-indexed)

    return {
        'mean': column.mean(),
        'median': column.median(),
        'std': column.std()
    }
```

```python
# Write a Pandas function to find the correlation between the first two columns.
def correlation():
    df = pd.DataFrame(iris_2d)
    return df[0].corr(df[1])
```

```python
# Write a Pandas function to create a new column for volume in iris_2d, where volume is (pi x petallength x sepal_length^2)/3
# Hints:
# 1. Introduce new dimension to match iris_2d's
# 2. Add the new column
def new_column():
    df = pd.DataFrame(iris_2d, columns=['sepal_length','sepal_width','petal_length','petal_width'])
    volume = (np.pi * df['petal_length'] * (df['sepal_length'] ** 2)) / 3
    volume_2d = volume.values.reshape(-1, 1)
    df['volume'] = volume_2d
```

# Q4 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
def bin_petal_length(arr):
    df = pd.DataFrame(arr, columns=['sepallength','sepalwidth','petallength','petalwidth'])
    bins = [0, 3, 5, np.inf]
    labels = ['small', 'medium', 'large']
    df['petal_length_bin'] = pd.cut(df['petallength'].astype(float), bins=bins, labels=labels, right=False)
    return df['petal_length_bin']
binned = bin_petal_length(iris_2d)
print(binned.head())
```

Q5 Write a Pandas function to find the most frequent value of petal length (3rd column) in iris dataset

```python
def most_frequent_value():
    df = pd.DataFrame(iris, columns=names)
    petal_length = df['petallength'].astype(float)
    return petal_length.mode()[0]
```
