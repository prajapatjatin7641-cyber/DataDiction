# Q4 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python
import pandas as pd
def bin_petal_length(iris_2d):
  petal_len=iris_2d.iloc[:, 2]
  bins = [float("-inf"), 3, 5, float("inf")]
  label=["small","medium","large"]
  return pd.cut(petal_len,bins=bins,labels=label).to_numpy
```
